<br><h1>Disulphide analysis</h1>

Here we will plot the annotated distances for every disulphide bond reported in PDB structures. You will need to patch the path in the CoorFileWalk to match your computer.

```python
import gemmi
from tqdm.notebook import tqdm
import numpy as np

disulphide_lengths = [ ]

for path in tqdm(gemmi.CoorFileWalk("/Users/jon/repos/pdb_mirror/data/structures/divided/mmCIF")) :
  structure = gemmi.read_structure(path)
  model = structure[0]
  conn_list = structure.connections
  for conn in conn_list :
    if conn.type == gemmi.ConnectionType.Disulf :
      disulphide_lengths.append(conn.reported_distance)
      if conn.reported_distance > 2.2 :
        print (path)
        print (conn)
        print (conn.reported_distance)

print ("Data collected for {:d} annotated disulphide bonds, " \
       "with an average of {:0.2f} and a standard deviation of {:0.2f}." \
       .format(len(disulphide_lengths), np.average(disulphide_lengths), \
               np.std(disulphide_lengths)))

plt.title ( "Disulphide bond lengths - high resolution dataset",
             fontsize = 15,
             fontweight = "bold" )
plt.hist (disulphide_lengths, bins=50)
plt.axvline(np.average(disulphide_lengths), color='orange', linestyle='dashed',\
            linewidth=3)
plt.axvline(np.average(disulphide_lengths) - np.std(disulphide_lengths), \
            color='orange', linestyle='dashed', linewidth=1)
plt.axvline(np.average(disulphide_lengths) + np.std(disulphide_lengths), \
            color='orange', linestyle='dashed', linewidth=1)
plt.xlabel("Length (A)")
plt.ylabel ("Frequency")
plt.show()
```
